```python
import numpy as np
import matplotlib.pyplot as plt

# Creating Data
x = np.arange(0,50)
y1 = np.random.rand(len(x)) # random numbers 
y2 = np.zeros_like(x)  # linear relation

for i in range(len(x)):
    y2[i] = (x[i]+np.random.normal(loc=5))
```

```python
# Training data
a = list(range(0,50,1))

x_train = np.zeros(30)
y1_train = np.zeros_like(x_train)
y2_train = np.zeros_like(x_train)


for i in range(30):
    rand_int = np.random.choice(a)
    a.pop(a.index(rand_int))
    
    x_train[i] = x[rand_int]
    y1_train[i] = y1[rand_int]
    y2_train[i] = y2[rand_int]
    
```

```python
def beta_1_hat(x, y):
    num = 0
    den = 0
    for i in range(len(x)):
        num += (x[i] - np.mean(x))*(y[i] - np.mean(y))
        den += (x[i] - np.mean(x))**2
    return num/den

def beta_0_hat(x,y):
    beta1 = beta_1_hat(x,y)
    return np.mean(y) - beta1*np.mean(x)
```

```python
b1 = beta_1_hat(x_train, y2_train)
b0 = beta_0_hat(x_train, y2_train)
print(b0,b1)
plt.plot(x_train, y2_train, 'ro')
plt.plot(x, b0+x*b1)
plt.show()


b1 = beta_1_hat(x_train, y1_train)
b0 = beta_0_hat(x_train, y1_train)
print(b0,b1)
plt.plot(x_train, y1_train, 'ro')
plt.plot(x, b0+x*b1)
```
