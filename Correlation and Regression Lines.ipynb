Here are the test scores of 10 students in physics and history:

Physics Scores  15  12  8   8   7   7   7   6   5   3
History Scores  10  25  17  11  13  17  20  13  9   15

Compute Karl Pearson’s coefficient of correlation between these scores.

Compute the answer correct to three decimal places.

Compute the slope of the line of regression obtained while treating Physics as the independent variable.
Compute the answer correct to three decimal places.

```python
PhysicsScore = np.array([15,12,8,8,7,7,7,6,5,3]) 
HistoryScore = np.array([10,25,17,11,13,17,20,13,8,15])


# Using linear regression to get the Pearson coefficent
PhysicsScore = PhysicsScore.reshape((-1,1))
reg = LinearRegression().fit(PhysicsScore, HistoryScore)

# plotting scatter plot of scores and regression fit
plt.scatter(PhysicsScore, HistoryScore, label='Students')
plt.plot(PhysicsScore, reg.predict(PhysicsScore), color='r', label='Regression')
plt.xlabel('Physics Score')
plt.ylabel('History Score')
plt.title('Linear Regression for 10 Students')
plt.legend(loc='upper left')
plt.show()
slope = reg.coef_
print('Slope of line of regression is: ' + str(slope)[1:6])
```

When a student scores 10 in Physics, what is his probable score in History? Compute the answer correct to one decimal place.

```python
PhysicsScore = np.array([15,12,8,8,7,7,7,6,5,3]) 
HistoryScore = np.array([10,25,17,11,13,17,20,13,8,15])


# Using linear regression to get the Pearson coefficent
PhysicsScore = PhysicsScore.reshape((-1,1))
reg = LinearRegression().fit(PhysicsScore, HistoryScore)
score = reg.predict([[10]])[0]
print('Probable Score in History: ' + str(score)[:4])
```

The two regression lines of a bivariate distribution are:
4x – 5y + 33 = 0 (line of y on x), 20x – 9y – 107 = 0 (line of x on y).
Estimate the value of x when  y= 7. Compute the correct answer to one decimal place.

```python
def x(y):
    return (9*y + 107)/20
def y(x):
    return (4*x + 33)/5


# 4x – 5y + 33 = 0 is the regression line of y regressed on x 
# 20x – 9y – 107 = 0 is the regression line of x regressed on y 
# since we are given y=7, we will use the regression line of x regressed on y.

print(str(x(7)))
```

The two regression lines of a bivariate distribution are:

4x – 5y + 33 = 0 (line of y on x), 20x – 9y – 107 = 0 (line of x on y).


Find the variance of y when $\sigma$x=3. Compute the correct answer to one decimal place.

```python
def x(y):
    return (9*y + 107)/20
def y(x):
    return (4*x + 33)/5

# Slope of x(y) is 9/20
# Slope of y(x) is 4/5

def VarY(slopeY, slopeX, VarX):
    # slopeY = slope when y is dependent variable
    # slopeX = slope when x is dependent variable
    # VarX = variance of x (3**2 in our case)
    return (slopeY/slopeX)*VarX
    
print(str(VarY(4/5,9/20,3**2)))
```
